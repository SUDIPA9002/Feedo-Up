### Importing modules

```python
import requests
from bs4 import BeautifulSoup
```

```python
link = 'https://www.flipkart.com/realme-11x-5g-purple-dawn-128-gb/product-reviews/itm07be1a2ff1a1b?pid=MOBGS2WFBDQBD2B4&lid=LSTMOBGS2WFBDQBD2B4HUEFK5&marketplace=FLIPKART&page={}'
```

### Header for request


```python
# User-Agent and Accept-Language headers
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36',
    'Accept-Language': 'en-us,en;q=0.5'
}
```

```python
# Number of pages to scrape (determine this from the website)
num_pages = 50
```

### Getting html and Parsing

```python
for page_number in range(1, num_pages + 1):
    # Construct the URL for the current page
    url = link.format(page_number)
    
    # Send a GET request to the page
    page = requests.get(url, headers=headers)
    
    # Parse the HTML content
    soup = BeautifulSoup(page.content, 'html.parser')
```

## Extract Customer names , Review title , Ratings & Comments

```python

customer_names = []
review_title = []
ratings = []
comments = []

for page_number in range(1, num_pages + 1):

    # Extract customer names
    names = soup.find_all('p', class_='_2sc7ZR _2V5EHH')
    for name in names:
        customer_names.append(name.get_text())
    
    # Extract review titles
    title = soup.find_all('p', class_='_2-N8zT')
    for t in title:
        review_title.append(t.get_text())
    
    # Extract ratings
    rat = soup.find_all('div', class_='_3LWZlK _1BLPMq')
    for r in rat:
        ratings.append(r.get_text())
    
    # Extract comments
    cmt = soup.find_all('div', class_='t-ZTKy')
    for c in cmt:
        comment_text = c.div.div.get_text(strip=True)
        comments.append(comment_text)
        
```

```python
print(len(customer_names))
print(len(review_title))
print(len(ratings))
print(len(comments))
```

## Create a DataFrame to store the data

```python
import pandas as pd
```

```python
data = {
    'Customer Name': customer_names,
    'Review Title': review_title,
    'Rating': ratings,
    'Comment': comments
}

df = pd.DataFrame(data)
```

```python
# Save the DataFrame to a CSV file
df.to_csv('realme 11x 5G (Purple Dawn, 128 GB).csv', index=False)
```
