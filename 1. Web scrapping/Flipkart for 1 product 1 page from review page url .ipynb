### Importing modules

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.flipkart.com/realme-11x-5g-purple-dawn-128-gb/product-reviews/itm07be1a2ff1a1b?pid=MOBGS2WFBDQBD2B4&lid=LSTMOBGS2WFBDQBD2B4HUEFK5&marketplace=FLIPKART&page=1'
```

### Header for request


```python
Header = ({'Useer-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36','Accept-Language':'en-us,en;q=0.5'})
```

### Getting html and Parsing

num_pages = 5

```python
page = requests.get(url,headers=Header)
```

```python
page
```

```python
# page.content
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
print(soup.prettify())
```

### Getting names of Customer

```python
names = soup.find_all('p',class_= '_2sc7ZR _2V5EHH')
```

```python
names
```

```python
customer_name = []
for i in range(0,len(names)):
    customer_name.append(names[i].get_text())
customer_name
```

### Comment about the product

```python
comment = soup.find_all('div',class_= 't-ZTKy')
```

```python
comment
```

```python
comments = []
for i in range(0,len(comment)):
    comment_text = comment[i].div.div.get_text(strip=True)
    comments.append(comment_text)

comments
```
